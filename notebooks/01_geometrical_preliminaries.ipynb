## 1.1 Geometry of Three-Dimensional Space

Physical problems in civil engineering — stress in a dam, flow in a pipe, deformation of a bridge — all unfold in three-dimensional space. Before we can describe them mathematically we need a precise vocabulary for the geometric objects that live there.

| Object | Dimension | Size | Key property |
|--------|-----------|------|--------------|
| **Point** | 0D | None | Position only |
| **Curve** $\mathcal{C}$ | 1D | Arc length | Orientation = tangent; rate of change of orientation = curvature |
| **Surface** $\mathcal{P}$ | 2D | Area | Orientation = normal to surface |
| **Body** $\mathcal{B}$ | 3D | Volume | Each interior point carries a physical field (temperature, stress, …) |

We set up a **coordinate system** — a reference frame — to assign numbers to points. The standard choice is a right-handed Cartesian frame with mutually orthogonal axes $x_1, x_2, x_3$.

## 1.2 What is a Vector?

```{admonition} Definition
:class: definition
A **vector** $\mathbf{v}$ is a directed line segment characterised by:
- **Magnitude** $\|\mathbf{v}\|$ — the length of the segment
- **Direction** — the orientation of the segment, indicated by an arrowhead
```

Physical examples that are vectors: force, displacement, velocity, acceleration, electric field.

```{admonition} Key Insight (highlighted in lecture)
:class: important
**The magnitude and direction of a vector are intrinsic physical properties — they are independent of the choice of coordinate system.** Only the *components* of a vector change when we change coordinates.
```

### Graphical representation

A vector $\mathbf{v}$ from point $a$ to point $b$ is drawn as an arrow. The *shaft* gives orientation; the *arrowhead* gives direction; the *length of the shaft* represents the magnitude $\|\mathbf{v}\|$.

Two vectors are **equal** if and only if they have the same magnitude and the same direction — their position in space does not matter (free vectors).


````python
"""Interactive: drag-style visualisation of a vector and its components.
   Uses matplotlib sliders to change the vector components and see the
   magnitude/direction update in real time."""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.widgets import Slider

fig, ax = plt.subplots(figsize=(9, 7))
plt.subplots_adjust(left=0.1, bottom=0.28)

v0 = np.array([3.0, 2.0])

def draw_vector(v):
    ax.cla()
    mag = np.linalg.norm(v)
    angle = np.degrees(np.arctan2(v[1], v[0]))
    # Grid
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.set_xlim(-5, 5); ax.set_ylim(-5, 5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.set_xlabel(r'$x_1$', fontsize=13); ax.set_ylabel(r'$x_2$', fontsize=13)

    # Component dashed lines
    ax.plot([0, v[0]], [0, 0], 'b--', linewidth=1.3, alpha=0.7)
    ax.plot([v[0], v[0]], [0, v[1]], 'r--', linewidth=1.3, alpha=0.7)
    ax.annotate('', xy=(v[0], 0), xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color='blue', lw=1.5))
    ax.annotate('', xy=(v[0], v[1]), xytext=(v[0], 0),
                arrowprops=dict(arrowstyle='->', color='red', lw=1.5))

    # Main vector
    ax.annotate('', xy=(v[0], v[1]), xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='black', lw=2.5))

    # Labels
    ax.text(v[0]/2, -0.35, rf'$v_1 = {v[0]:.2f}$', ha='center', color='blue', fontsize=11)
    ax.text(v[0]+0.25, v[1]/2, rf'$v_2 = {v[1]:.2f}$', ha='left', color='red', fontsize=11)
    ax.text(v[0]/2 - 0.4, v[1]/2 + 0.3, r'$\mathbf{v}$', fontsize=16, fontweight='bold')
    ax.set_title(
        rf'$\mathbf{{v}} = ({v[0]:.2f},\; {v[1]:.2f})$  |  '
        rf'$\|\mathbf{{v}}\| = {mag:.3f}$  |  '
        rf'$\theta = {angle:.1f}^\circ$',
        fontsize=12)
    # Angle arc
    theta_arc = np.linspace(0, np.arctan2(v[1], v[0]), 60)
    r_arc = 0.7
    ax.plot(r_arc*np.cos(theta_arc), r_arc*np.sin(theta_arc), 'k', linewidth=1)
    mid = len(theta_arc)//2
    ax.text(1.0*np.cos(theta_arc[mid]), 1.0*np.sin(theta_arc[mid]), r'$\theta$', fontsize=12)
    fig.canvas.draw_idle()

draw_vector(v0)

# Sliders
ax_v1 = plt.axes([0.15, 0.15, 0.70, 0.03])
ax_v2 = plt.axes([0.15, 0.08, 0.70, 0.03])
s_v1 = Slider(ax_v1, r'$v_1$', -4.5, 4.5, valinit=v0[0], valstep=0.1)
s_v2 = Slider(ax_v2, r'$v_2$', -4.5, 4.5, valinit=v0[1], valstep=0.1)

def update(val):
    draw_vector(np.array([s_v1.val, s_v2.val]))

s_v1.on_changed(update)
s_v2.on_changed(update)
plt.suptitle('Interactive: Vector components, magnitude, and angle', y=0.97, fontsize=12, fontweight='bold')
plt.show()
````

## 1.3 Basis Vectors and Component Representation

To work with vectors algebraically we choose an **orthonormal basis** — a set of three mutually perpendicular unit vectors $\{\hat{\mathbf{e}}_1, \hat{\mathbf{e}}_2, \hat{\mathbf{e}}_3\}$ aligned with the coordinate axes.

$$
\hat{\mathbf{e}}_i \cdot \hat{\mathbf{e}}_j = \delta_{ij}, \qquad \|\hat{\mathbf{e}}_i\| = 1
$$

Any vector can be written as a linear combination of the basis vectors:

$$
\mathbf{v} = v_1 \hat{\mathbf{e}}_1 + v_2 \hat{\mathbf{e}}_2 + v_3 \hat{\mathbf{e}}_3 = \sum_{i=1}^{3} v_i \hat{\mathbf{e}}_i
$$

The scalars $v_1, v_2, v_3$ are the **components** of $\mathbf{v}$ in the basis $\{\hat{\mathbf{e}}_i\}$. Crucially:

$$
v_i = \mathbf{v} \cdot \hat{\mathbf{e}}_i \quad \text{(projection onto the }i\text{-th axis)}
$$

The **magnitude** (Euclidean norm) is:

$$
\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + v_3^2}
$$

A **unit vector** in the direction of $\mathbf{v}$ is:

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}, \qquad \|\hat{\mathbf{v}}\| = 1
$$


````python
"""Symbolic computation: vector in a basis."""
import sympy as sp

# Define symbolic components
v1, v2, v3 = sp.symbols('v_1 v_2 v_3', real=True)

v = sp.Matrix([v1, v2, v3])
mag_sq = v.dot(v)
mag = sp.sqrt(mag_sq)
v_unit = v / mag

print("Vector v:", v.T)
print("Magnitude ||v||:", mag)
print()

# Numerical example
import numpy as np
vals = {v1: 3, v2: 4, v3: 0}
v_num = np.array([3., 4., 0.])
print("Example: v = (3, 4, 0)")
print(f"  ||v|| = {np.linalg.norm(v_num):.4f}  (should be 5.0)")
print(f"  v_hat = {v_num / np.linalg.norm(v_num)}")
print(f"  ||v_hat|| = {np.linalg.norm(v_num / np.linalg.norm(v_num)):.6f}")
````

## 1.4 Summary

| Concept | Symbol | Formula |
|---------|--------|---------|
| Vector | $\mathbf{v}$ | Directed line segment |
| Component | $v_i$ | $\mathbf{v}\cdot\hat{\mathbf{e}}_i$ |
| Magnitude | $\|\mathbf{v}\|$ | $\sqrt{v_i v_i}$ |
| Unit vector | $\hat{\mathbf{v}}$ | $\mathbf{v}/\|\mathbf{v}\|$ |
| Zero vector | $\mathbf{0}$ | $\|\mathbf{0}\| = 0$ |
