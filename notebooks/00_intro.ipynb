Welcome to the interactive lecture notes for **CV 1050**. These notes are designed to complement the in-class chalk-and-talk sessions. Each chapter contains:

- 📖 **Theory** — concise, rigorous exposition matched to the lecture slides
- 🖊️ **Worked examples** — step-by-step derivations with symbolic computation
- 🎛️ **Interactive widgets** — sliders and plots to build geometric intuition
- 💻 **AI-guided coding** — structured tasks where you write and verify code yourself
- ✏️ **Self-assessment problems** — pen-and-paper exercises that build exam-readiness

## How to use these notes

```{admonition} For Students
:class: tip
- Read through the theory **before** each lecture as a primer.
- Attend the chalk-and-talk session; take your own notes.
- Return here after the lecture to run the interactive examples.
- Complete the **Self-Assessment** problems on paper — these are your exam preparation.
- In the **Coding Tasks**, you are expected to write the code yourself. Using AI to explain a concept is encouraged. Submitting AI-generated code as your own is academic dishonesty.
```

```{admonition} On the use of AI tools
:class: warning
AI tools (ChatGPT, Claude, Copilot, etc.) are permitted **as a tutor**, not as a coder-for-hire. Use AI to:
- Ask *why* a formula works
- Get a different explanation of a concept
- Debug an error message

Do **not** ask AI to write your assignment code and submit it. Assessments are designed so that correct answers require understanding demonstrated through in-class tests, oral vivas, and derive-then-implement tasks.
```
