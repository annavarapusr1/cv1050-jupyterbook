## 6.1 Tensors as Linear Maps

A **second-order tensor** $\\mathbf{T}$ is a linear map from vectors to vectors:

$$\\mathbf{T}\\mathbf{u} = \\mathbf{v} \\qquad \\text{(linearity: } \\mathbf{T}(\\alpha\\mathbf{u}+\\beta\\mathbf{w}) = \\alpha\\mathbf{T}\\mathbf{u}+\\beta\\mathbf{T}\\mathbf{w}\\text{)}$$

Physical examples: **stress tensor** (normal -> traction), **strain tensor**, **inertia tensor**.

## 6.2 Tensor Product (Dyad)

For vectors $\\mathbf{u}$ and $\\mathbf{v}$, the dyad $\\mathbf{u}\\otimes\\mathbf{v}$ is:

$$[\\mathbf{u}\\otimes\\mathbf{v}]\\mathbf{w} = (\\mathbf{v}\\cdot\\mathbf{w})\\mathbf{u}, \\qquad [\\mathbf{u}\\otimes\\mathbf{v}]_{ij} = u_i v_j$$

The 9 dyads $\\hat{\\mathbf{e}}_i\\otimes\\hat{\\mathbf{e}}_j$ are a basis: any second-order tensor is $\\mathbf{T} = T_{ij}[\\hat{\\mathbf{e}}_i\\otimes\\hat{\\mathbf{e}}_j]$.

## 6.3 Key Operations

| Operation | Index form | Result |
|-----------|-----------|--------|
| Tensor-vector product | $[\\mathbf{T}\\mathbf{u}]_i = T_{ij}u_j$ | vector |
| Tensor-tensor product | $[\\mathbf{TS}]_{ij} = T_{ik}S_{kj}$ | tensor |
| Transpose | $[\\mathbf{T}^T]_{ij} = T_{ji}$ | tensor |
| Trace | $\\text{tr}(\\mathbf{T}) = T_{ii}$ | scalar |
| Double contraction | $\\mathbf{T}:\\mathbf{S} = T_{ij}S_{ij}$ | scalar |
| Identity | $I_{ij} = \\delta_{ij}$ | $\\mathbf{I}\\mathbf{u}=\\mathbf{u}$ |

## 6.4 Symmetry

- **Symmetric**: $T_{ij}=T_{ji}$ (6 independent components)
- **Skew-symmetric**: $T_{ij}=-T_{ji}$ (3 independent components; $T_{ii}=0$)
- Any tensor: $\\mathbf{T} = \\frac{1}{2}(\\mathbf{T}+\\mathbf{T}^T) + \\frac{1}{2}(\\mathbf{T}-\\mathbf{T}^T)$

## 6.5 Tensor Inverse

$\\mathbf{T}^{-1}$ exists iff $\\det(\\mathbf{T})\\neq 0$: $\\mathbf{T}\\mathbf{T}^{-1}=\\mathbf{T}^{-1}\\mathbf{T}=\\mathbf{I}$

```python
# Numerical verification of tensor operations
import numpy as np

T = np.array([[2., 1., 0.],
              [1., 3., -1.],
              [0., -1., 4.]])
S = np.array([[1., 0., 2.],
              [0., 2., 1.],
              [2., 1., 3.]])
u = np.array([1., 2., -1.])

print("T ="); print(T)
print("\nT*u =", T @ u)
print("tr(T) =", np.trace(T))
print("det(T) =", round(np.linalg.det(T), 4))
print("T:S (double contraction) =", np.einsum('ij,ij->', T, S))
print("T is symmetric:", np.allclose(T, T.T))

# Dyad: u tensor v
v = np.array([0., 1., 2.])
dyad = np.outer(u, v)
w = np.array([1., -1., 0.5])
print("\n(u tensor v)*w =", dyad @ w)
print("(v.w)*u =", np.dot(v,w) * u)
print("Match:", np.allclose(dyad @ w, np.dot(v,w)*u))
```

```python
# Interactive: tensor as linear map on unit circle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plt.subplots_adjust(bottom=0.30)

def draw(T11, T12, T21, T22):
    T = np.array([[T11, T12],[T21, T22]])
    theta = np.linspace(0, 2*np.pi, 300)
    circle = np.array([np.cos(theta), np.sin(theta)])
    ellipse = T @ circle
    det_T = np.linalg.det(T)
    sym = np.allclose(T, T.T)

    for ax, pts, title, col in [
        (axes[0], circle, 'Input: unit circle', 'royalblue'),
        (axes[1], ellipse, f'T(circle) | det={det_T:.2f} | sym={sym}', 'firebrick')]:
        ax.cla()
        ax.plot(pts[0], pts[1], color=col, lw=2)
        ax.fill(pts[0], pts[1], alpha=0.15, color=col)
        ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
        ax.axhline(0,color='gray',lw=0.5); ax.axvline(0,color='gray',lw=0.5)
        lim = max(2., np.max(np.abs(pts))*1.2)
        ax.set_xlim(-lim,lim); ax.set_ylim(-lim,lim)
        ax.set_title(title, fontsize=10, fontweight='bold')

    # Draw image of basis vectors
    for vec, lbl in [(T[:,0], r'$T\hat{e}_1$'), (T[:,1], r'$T\hat{e}_2$')]:
        axes[1].annotate('', xy=vec, xytext=(0,0),
                         arrowprops=dict(arrowstyle='->', color='darkgreen', lw=2))
        axes[1].text(*(vec*1.1), lbl, color='darkgreen', fontsize=11)
    fig.canvas.draw_idle()

draw(2., 0.5, 0.5, 1.)
ax_s = [plt.axes([0.10+0.45*(i//2), 0.15+0.07*(i%2), 0.30, 0.03]) for i in range(4)]
labels = ['$T_{11}$','$T_{21}$','$T_{12}$','$T_{22}$']
inits  = [2., 0.5, 0.5, 1.]
sliders = [Slider(a, lb, -2.5, 2.5, valinit=iv, valstep=0.1)
           for a, lb, iv in zip(ax_s, labels, inits)]
def upd(v): draw(sliders[0].val, sliders[2].val, sliders[1].val, sliders[3].val)
for s in sliders: s.on_changed(upd)
plt.suptitle('Tensor as a linear map: deformation of the unit circle', fontsize=12, fontweight='bold')
plt.show()
```
