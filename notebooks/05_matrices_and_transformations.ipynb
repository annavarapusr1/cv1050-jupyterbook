## 5.1 Matrix Representation of Linear Maps

A linear map $\mathbf{A}: \mathbb{R}^3 \to \mathbb{R}^3$ in component form:

$$[\mathbf{A}\mathbf{v}]_i = A_{ij} v_j \qquad \text{(matrix-vector product)}$$

$$[\mathbf{A}\mathbf{B}]_{ij} = A_{ik} B_{kj} \qquad \text{(matrix-matrix product)}$$

$$[\mathbf{A}^T]_{ij} = A_{ji} \qquad (\mathbf{A}\mathbf{B})^T = \mathbf{B}^T\mathbf{A}^T$$

## 5.2 Coordinate Transformation

Given two Cartesian frames, the direction cosines $Q_{ij} = \hat{\mathbf{e}}'_i \cdot \hat{\mathbf{e}}_j$ define the rotation matrix $[Q]$.

**Vector transform:** $v'_i = Q_{ij} v_j$

**Tensor transform:** $T'_{ij} = Q_{ik} Q_{jl} T_{kl}$, i.e., $[T'] = [Q][T][Q]^T$

**Orthogonality:** $[Q][Q]^T = [I]$, $\det[Q] = \pm 1$


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Layout

def plot_rotation(v1=2.0, v2=1.0, theta_deg=30.0):
    theta = np.radians(theta_deg)
    Q = np.array([[ np.cos(theta), np.sin(theta)],
                  [-np.sin(theta), np.cos(theta)]])
    v = np.array([v1, v2])
    vp = Q @ v
    e1p = np.array([ np.cos(theta), np.sin(theta)])
    e2p = np.array([-np.sin(theta), np.cos(theta)])

    fig, ax = plt.subplots(figsize=(8, 7))
    ax.set_aspect('equal')
    lim = max(np.linalg.norm(v), 1.0) + 0.8
    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
    ax.grid(True, alpha=0.3)
    aw = dict(head_width=0.07, head_length=0.07, length_includes_head=True)

    for d, lbl, c in zip([[1,0],[0,1]], [r'$\hat{e}_1$',r'$\hat{e}_2$'], ['#C0392B','#27AE60']):
        ax.arrow(0,0,d[0],d[1], color=c, **aw)
        ax.text(d[0]*1.15, d[1]*1.15, lbl, color=c, fontsize=13, fontweight='bold')
    for ep, lbl, c in zip([e1p,e2p], [r"$\hat{e}'_1$",r"$\hat{e}'_2$"], ['#E74C3C','#2ECC71']):
        ax.arrow(0,0,ep[0],ep[1], color=c, **aw)
        ax.text(ep[0]*1.2, ep[1]*1.2, lbl, color=c, fontsize=13, fontweight='bold')
    ax.arrow(0,0,v[0],v[1], color='#8E44AD', head_width=0.1, head_length=0.1,
             length_includes_head=True, lw=2.5)
    ax.text(v[0]+0.05, v[1]+0.05, r'$\mathbf{v}$', color='#8E44AD', fontsize=14, fontweight='bold')
    ax.plot([v[0], vp[0]*e1p[0]], [v[1], vp[0]*e1p[1]], ':', color='gray')
    ax.plot([v[0], vp[1]*e2p[0]], [v[1], vp[1]*e2p[1]], ':', color='gray')
    ax.set_title(fr'Rotation $\theta={theta_deg:.0f}°$  |  v=({v[0]:.2f},{v[1]:.2f})  '
                 fr'v'"'"r'=({vp[0]:.2f},{vp[1]:.2f})  ||v||={np.linalg.norm(v):.3f}', fontsize=10)
    ax.set_xlabel(r'$x_1$'); ax.set_ylabel(r'$x_2$')
    plt.tight_layout()
    plt.show()

interact(plot_rotation,
         v1=FloatSlider(2.0, min=-3., max=3., step=0.1, description=r'$v_1$'),
         v2=FloatSlider(1.0, min=-3., max=3., step=0.1, description=r'$v_2$'),
         theta_deg=FloatSlider(30., min=0., max=360., step=5., description=r'theta'));
```

```python
import numpy as np

def R3z(deg):
    t = np.radians(deg)
    return np.array([[np.cos(t), np.sin(t), 0],
                     [-np.sin(t), np.cos(t), 0],
                     [0, 0, 1]])

T = np.array([[10.,5.,0.],[5.,20.,0.],[0.,0.,15.]])
Q = R3z(45)
Tp = Q @ T @ Q.T

print("Stress [T] (MPa):"); print(T)
print("\nRotated [T'] (MPa):"); print(np.round(Tp, 4))
print(f"\ntr(T)={np.trace(T):.1f}  tr(T')={np.trace(Tp):.6f}  <- invariant")
print(f"det(T)={np.linalg.det(T):.2f}  det(T')={np.linalg.det(Tp):.6f}  <- invariant")
print(f"Q orthogonal: {np.allclose(Q@Q.T, np.eye(3))}  det(Q)={np.linalg.det(Q):.6f}")
```
