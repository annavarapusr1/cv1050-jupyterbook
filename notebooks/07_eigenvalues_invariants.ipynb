## 7.1 Eigenvalue Problem

An **eigenvector** $\\hat{\\mathbf{n}}$ of a tensor $\\mathbf{T}$ is a special direction that is preserved (only scaled) by $\\mathbf{T}$:

$$\\mathbf{T}\\hat{\\mathbf{n}} = \\lambda\\hat{\\mathbf{n}}$$

The scalar $\\lambda$ is the corresponding **eigenvalue**. The equation says: acting on $\\hat{\\mathbf{n}}$ with $\\mathbf{T}$ gives back $\\hat{\\mathbf{n}}$ scaled by $\\lambda$.

## 7.2 Finding Eigenvalues: Characteristic Equation

Rearranging: $(\\mathbf{T} - \\lambda\\mathbf{I})\\hat{\\mathbf{n}} = \\mathbf{0}$

Non-trivial solutions exist iff the matrix $(\\mathbf{T}-\\lambda\\mathbf{I})$ is **singular**:

$$\\det(\\mathbf{T} - \\lambda\\mathbf{I}) = 0$$

Expanding this $3\\times3$ determinant gives the **characteristic equation**:

$$-\\lambda^3 + I_T\\lambda^2 - II_T\\lambda + III_T = 0$$

where the coefficients are the **principal invariants** of $\\mathbf{T}$:

| Invariant | Formula | Physical meaning |
|-----------|---------|-----------------|
| $I_T$ | $\\text{tr}(\\mathbf{T}) = T_{ii}$ | Sum of eigenvalues |
| $II_T$ | $\\frac{1}{2}(I_T^2 - \\text{tr}(\\mathbf{T}^2)) = \\frac{1}{2}(T_{ii}^2 - T_{ij}T_{ij})$ | Sum of products of pairs of eigenvalues |
| $III_T$ | $\\det(\\mathbf{T}) = \\varepsilon_{ijk}T_{1i}T_{2j}T_{3k}$ | Product of all eigenvalues |

The characteristic equation has 3 roots (eigenvalues). For symmetric tensors all eigenvalues are **real**.

## 7.3 Spectral Decomposition

For a **symmetric** tensor $\\mathbf{T}$ with eigenvalues $\\lambda_1, \\lambda_2, \\lambda_3$ and orthonormal eigenvectors $\\hat{\\mathbf{n}}_1, \\hat{\\mathbf{n}}_2, \\hat{\\mathbf{n}}_3$:

$$\\boxed{\\mathbf{T} = \\sum_{i=1}^{3} \\lambda_i\\,(\\hat{\\mathbf{n}}_i \\otimes \\hat{\\mathbf{n}}_i)}$$

This is the **spectral representation** — the tensor is expressed in its *own* preferred basis (the principal axes).

```{admonition} Key insight
:class: important
In the eigenvector basis, the matrix of $\\mathbf{T}$ is **diagonal**: $[\\mathbf{T}]_{\\hat{n}} = \\text{diag}(\\lambda_1, \\lambda_2, \\lambda_3)$.
The principal invariants are the coefficients of this characteristic polynomial and are **frame-independent**.
```

## 7.4 Fundamental vs Principal Invariants

The notes distinguish two sets:
- **Principal invariants**: $I_T, II_T, III_T$ (coefficients of characteristic polynomial)
- **Fundamental invariants**: $J_1 = \\text{tr}(\\mathbf{T})$, $J_2 = \\text{tr}(\\mathbf{T}^2)$, $J_3 = \\text{tr}(\\mathbf{T}^3)$

They are related but not identical:

$$I_T = J_1, \\qquad II_T = \\tfrac{1}{2}(J_1^2 - J_2), \\qquad III_T = \\tfrac{1}{6}(J_1^3 - 3J_1J_2 + 2J_3)$$

## 7.5 Cayley-Hamilton Theorem

```{admonition} Theorem (Cayley-Hamilton)
:class: important
Every matrix satisfies its own characteristic equation:
$$-\\mathbf{T}^3 + I_T\\mathbf{T}^2 - II_T\\mathbf{T} + III_T\\mathbf{I} = \\mathbf{0}$$
```

This has a powerful consequence: $\\mathbf{T}^3$ can always be expressed in terms of $\\mathbf{I}$, $\\mathbf{T}$, and $\\mathbf{T}^2$ — so any polynomial in $\\mathbf{T}$ reduces to at most a quadratic.

````python
# Eigenvalue problem: step-by-step numerical + verification
import numpy as np

T = np.array([[3., 1., 0.],
              [1., 2., 1.],
              [0., 1., 4.]])

print("T ="); print(T)
print(f"\nIs T symmetric: {np.allclose(T, T.T)}")

# Compute invariants
I_T   = np.trace(T)
II_T  = 0.5*(np.trace(T)**2 - np.trace(T@T))
III_T = np.linalg.det(T)
print(f"\nPrincipal invariants:")
print(f"  I_T  = tr(T)       = {I_T:.4f}")
print(f"  II_T = (I^2-tr(T^2))/2 = {II_T:.4f}")
print(f"  III_T = det(T)     = {III_T:.4f}")

# Characteristic polynomial: -lambda^3 + I*lambda^2 - II*lambda + III = 0
# Equivalently: lambda^3 - I*lambda^2 + II*lambda - III = 0
coeffs = [1., -I_T, II_T, -III_T]
lam_roots = np.roots(coeffs)
print(f"\nEigenvalues from characteristic polynomial: {np.sort(lam_roots.real)}")

# NumPy eigenvalues (for comparison)
lam_np, vec_np = np.linalg.eigh(T)
print(f"Eigenvalues from numpy.linalg.eigh:         {np.sort(lam_np)}")

# Verify Cayley-Hamilton
I_mat = np.eye(3)
CH = -T@T@T + I_T*(T@T) - II_T*T + III_T*I_mat
print(f"\nCayley-Hamilton residual ||...|| = {np.linalg.norm(CH):.2e}  (should be ~0)")

# Spectral decomposition
print(f"\nSpectral decomposition T = sum lambda_i (n_i x n_i):")
T_reconstructed = sum(lam_np[i]*np.outer(vec_np[:,i], vec_np[:,i]) for i in range(3))
print(f"Reconstruction error: {np.linalg.norm(T - T_reconstructed):.2e}")
````

````python
# Interactive: eigenvalue visualisation for a 2x2 symmetric matrix
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig, ax = plt.subplots(figsize=(8, 7))
plt.subplots_adjust(bottom=0.30)

def draw(T11, T12, T22):
    T = np.array([[T11, T12],[T12, T22]])
    lam, vec = np.linalg.eigh(T)
    I_T = np.trace(T); II_T = np.linalg.det(T)

    # Unit circle -> ellipse
    theta_arc = np.linspace(0, 2*np.pi, 300)
    circle = np.array([np.cos(theta_arc), np.sin(theta_arc)])
    ellipse = T @ circle

    ax.cla()
    ax.plot(ellipse[0], ellipse[1], 'royalblue', lw=1.5, alpha=0.7, label='T(unit circle)')
    ax.plot(circle[0], circle[1], 'k--', lw=1, alpha=0.4, label='unit circle')
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
    ax.axhline(0,color='gray',lw=0.5); ax.axvline(0,color='gray',lw=0.5)
    lim = max(2., max(abs(lam))*1.4)
    ax.set_xlim(-lim,lim); ax.set_ylim(-lim,lim)

    # Eigenvectors (scaled by eigenvalue)
    colors = ['firebrick','darkgreen']
    for i in range(2):
        ev = lam[i]*vec[:,i]
        ax.annotate('', xy=ev, xytext=(0,0),
                    arrowprops=dict(arrowstyle='->', color=colors[i], lw=2.5))
        ax.text(ev[0]+0.05, ev[1]+0.08,
                rf'$\lambda_{i+1}={lam[i]:.2f}$', color=colors[i], fontsize=11, fontweight='bold')

    info = (f"T = [[{T11:.1f},{T12:.1f}],[{T12:.1f},{T22:.1f}]]\n"
            f"$I_T = tr(T) = {I_T:.2f}$  |  $II_T = det(T) = {II_T:.2f}$\n"
            rf"$\lambda_1={lam[0]:.3f}$, $\lambda_2={lam[1]:.3f}$\n"
            rf"$\lambda_1+\lambda_2={lam[0]+lam[1]:.3f}$ (= $I_T$)  "
            rf"$\lambda_1\lambda_2={lam[0]*lam[1]:.3f}$ (= $II_T$)")
    ax.text(-lim+0.1, -lim+0.1, info, fontsize=9, va='bottom',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))
    ax.legend(loc='upper right', fontsize=9)
    ax.set_title('Symmetric tensor: eigenvectors = principal axes', fontsize=11, fontweight='bold')
    fig.canvas.draw_idle()

draw(3., 1., 2.)
sl_axes = [plt.axes([0.15, 0.20, 0.25, 0.03]),
           plt.axes([0.45, 0.20, 0.25, 0.03]),
           plt.axes([0.72, 0.20, 0.25, 0.03])]
labels = ['$T_{11}$','$T_{12}=T_{21}$','$T_{22}$']
inits  = [3., 1., 2.]
sliders = [Slider(a, lb, -3., 4., valinit=iv, valstep=0.25)
           for a, lb, iv in zip(sl_axes, labels, inits)]
def upd(v): draw(sliders[0].val, sliders[1].val, sliders[2].val)
for s in sliders: s.on_changed(upd)
plt.show()
````

````python
# Verify principal invariants are frame-independent
import numpy as np

def invariants(T):
    I   = np.trace(T)
    II  = 0.5*(np.trace(T)**2 - np.trace(T@T))
    III = np.linalg.det(T)
    return I, II, III

def rotation_3d(axis, angle_deg):
    theta = np.radians(angle_deg)
    n = np.array(axis, dtype=float); n /= np.linalg.norm(n)
    c, s = np.cos(theta), np.sin(theta)
    K = np.array([[0,-n[2],n[1]],[n[2],0,-n[0]],[-n[1],n[0],0]])
    return np.eye(3) + s*K + (1-c)*(K@K)

T = np.array([[3.,1.,0.],[1.,2.,1.],[0.,1.,4.]])
I0, II0, III0 = invariants(T)
print("Original frame invariants:")
print(f"  I={I0:.4f},  II={II0:.4f},  III={III0:.4f}")

for angle, axis in [(30,[0,0,1]), (45,[1,1,0]), (72,[1,2,3])]:
    Q = rotation_3d(axis, angle)
    Tp = Q @ T @ Q.T  # transformed tensor
    I, II, III = invariants(Tp)
    ok = all(abs(x-y) < 1e-8 for x,y in [(I,I0),(II,II0),(III,III0)])
    print(f"  Rotation {angle}deg about {axis}: I={I:.4f}, II={II:.4f}, III={III:.4f}  MATCH={ok}")
````
