## 2.1 Vector Addition

### Geometric rules

There are two equivalent geometric constructions:

**Head-to-Tail rule** — Place the tail of $\mathbf{v}$ at the head of $\mathbf{u}$. The sum $\mathbf{u}+\mathbf{v}$ is the vector from the tail of $\mathbf{u}$ to the head of $\mathbf{v}$.

**Parallelogram rule** — Place $\mathbf{u}$ and $\mathbf{v}$ tail-to-tail to form two sides of a parallelogram. The diagonal from the shared tail is $\mathbf{u}+\mathbf{v}$.

### Algebraic rule

Component-wise:

$$
\mathbf{u} + \mathbf{v} = (u_1+v_1)\hat{\mathbf{e}}_1 + (u_2+v_2)\hat{\mathbf{e}}_2 + (u_3+v_3)\hat{\mathbf{e}}_3
$$

### Properties (axioms of a vector space)

| Property | Statement |
|----------|-----------|
| Commutativity | $\mathbf{u}+\mathbf{v} = \mathbf{v}+\mathbf{u}$ |
| Associativity | $(\mathbf{u}+\mathbf{v})+\mathbf{w} = \mathbf{u}+(\mathbf{v}+\mathbf{w})$ |
| Additive identity | $\mathbf{u}+\mathbf{0} = \mathbf{u}$ |
| Additive inverse | $\mathbf{u}+(-\mathbf{u}) = \mathbf{0}$ |

## 2.2 Scalar Multiplication

For scalar $\alpha \in \mathbb{R}$:

$$
\alpha\,\mathbf{v} = \alpha v_1\hat{\mathbf{e}}_1 + \alpha v_2\hat{\mathbf{e}}_2 + \alpha v_3\hat{\mathbf{e}}_3
$$

- $\alpha > 0$: same direction, magnitude scaled by $\alpha$
- $\alpha < 0$: reversed direction, magnitude scaled by $|\alpha|$  
- $\alpha = 0$: zero vector $\mathbf{0}$

### Properties:

$$
(\alpha\beta)\mathbf{u} = \alpha(\beta\mathbf{u}), \qquad
(\alpha+\beta)\mathbf{u} = \alpha\mathbf{u}+\beta\mathbf{u}, \qquad
\alpha(\mathbf{u}+\mathbf{v}) = \alpha\mathbf{u}+\alpha\mathbf{v}
$$

````python
"""Interactive: Vector addition — head-to-tail and parallelogram rules."""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig, axes = plt.subplots(1, 2, figsize=(13, 6))
plt.subplots_adjust(bottom=0.32, wspace=0.3)

def draw_addition(u, v):
    w = u + v
    lim = max(np.linalg.norm(w) + 1, 3)

    # --- Head-to-tail ---
    ax = axes[0]; ax.cla()
    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
    ax.axhline(0,color='gray',lw=0.5); ax.axvline(0,color='gray',lw=0.5)
    ax.set_title('Head-to-Tail Rule', fontsize=12, fontweight='bold')
    kw = dict(arrowprops=dict(arrowstyle='->', lw=2.2))
    ax.annotate('', xy=u, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='royalblue', lw=2.2))
    ax.text(u[0]/2-0.2, u[1]/2+0.15, r'$\mathbf{u}$', color='royalblue', fontsize=13, fontweight='bold')
    ax.annotate('', xy=u+v, xytext=u, arrowprops=dict(arrowstyle='->', color='firebrick', lw=2.2))
    ax.text(u[0]+v[0]/2-0.15, u[1]+v[1]/2+0.15, r'$\mathbf{v}$', color='firebrick', fontsize=13, fontweight='bold')
    ax.annotate('', xy=w, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='darkgreen', lw=2.5, linestyle='dashed'))
    ax.text(w[0]/2+0.1, w[1]/2-0.3, r'$\mathbf{u}+\mathbf{v}$', color='darkgreen', fontsize=12, fontweight='bold')

    # --- Parallelogram ---
    ax = axes[1]; ax.cla()
    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
    ax.axhline(0,color='gray',lw=0.5); ax.axvline(0,color='gray',lw=0.5)
    ax.set_title('Parallelogram Rule', fontsize=12, fontweight='bold')
    # Shaded parallelogram
    para = plt.Polygon([[0,0], u, u+v, v], closed=True, alpha=0.12, color='gold', linewidth=0)
    ax.add_patch(para)
    # u and v from origin
    ax.annotate('', xy=u, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='royalblue', lw=2.2))
    ax.text(u[0]/2-0.2, u[1]/2+0.15, r'$\mathbf{u}$', color='royalblue', fontsize=13, fontweight='bold')
    ax.annotate('', xy=v, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='firebrick', lw=2.2))
    ax.text(v[0]/2-0.2, v[1]/2+0.15, r'$\mathbf{v}$', color='firebrick', fontsize=13, fontweight='bold')
    # Far edges (dashed)
    ax.plot([v[0], w[0]], [v[1], w[1]], '--', color='royalblue', lw=1.5, alpha=0.6)
    ax.plot([u[0], w[0]], [u[1], w[1]], '--', color='firebrick', lw=1.5, alpha=0.6)
    # Diagonal = sum
    ax.annotate('', xy=w, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='darkgreen', lw=2.5))
    ax.text(w[0]/2+0.1, w[1]/2-0.3, r'$\mathbf{u}+\mathbf{v}$', color='darkgreen', fontsize=12, fontweight='bold')

    # Difference vector
    diff = u - v
    ax.annotate('', xy=diff, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='purple', lw=1.8, linestyle='dotted'))
    ax.text(diff[0]/2+0.1, diff[1]/2+0.1, r'$\mathbf{u}-\mathbf{v}$', color='purple', fontsize=10)

    fig.canvas.draw_idle()

u0 = np.array([2., 1.]); v0 = np.array([1., 2.])
draw_addition(u0, v0)

ax_u1 = plt.axes([0.12, 0.20, 0.35, 0.025])
ax_u2 = plt.axes([0.12, 0.14, 0.35, 0.025])
ax_v1 = plt.axes([0.55, 0.20, 0.35, 0.025])
ax_v2 = plt.axes([0.55, 0.14, 0.35, 0.025])
su1 = Slider(ax_u1, r'$u_1$', -3., 3., valinit=2., valstep=0.25)
su2 = Slider(ax_u2, r'$u_2$', -3., 3., valinit=1., valstep=0.25)
sv1 = Slider(ax_v1, r'$v_1$', -3., 3., valinit=1., valstep=0.25)
sv2 = Slider(ax_v2, r'$v_2$', -3., 3., valinit=2., valstep=0.25)

def update(val):
    draw_addition(np.array([su1.val, su2.val]), np.array([sv1.val, sv2.val]))

for s in [su1, su2, sv1, sv2]:
    s.on_changed(update)

plt.suptitle('Vector Addition — use sliders to explore', fontsize=12, fontweight='bold', y=0.98)
plt.show()
````

## 2.3 Linear Combinations & Linear Independence

A **linear combination** of vectors $\mathbf{u}_1, \dots, \mathbf{u}_k$ is:

$$
\mathbf{w} = \alpha_1 \mathbf{u}_1 + \alpha_2 \mathbf{u}_2 + \cdots + \alpha_k \mathbf{u}_k
$$

The vectors $\{\mathbf{u}_i\}$ are **linearly independent** if the only solution to 
$\alpha_1\mathbf{u}_1 + \cdots + \alpha_k\mathbf{u}_k = \mathbf{0}$ is $\alpha_i = 0$ for all $i$.

The three basis vectors $\{\hat{\mathbf{e}}_1, \hat{\mathbf{e}}_2, \hat{\mathbf{e}}_3\}$ are linearly independent and span all of $\mathbb{R}^3$ — they form a **basis**.


````python
"""Visualise 3D vector addition and scalar scaling."""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 5))

# --- Left: 3D vector addition ---
ax1 = fig.add_subplot(121, projection='3d')
u3 = np.array([2., 1., 0.5])
v3 = np.array([0.5, 1.5, 1.])
w3 = u3 + v3

def arrow3(ax, start, end, color, label):
    d = end - start
    ax.quiver(*start, *d, color=color, linewidth=2, arrow_length_ratio=0.15)
    mid = start + 0.5*d
    ax.text(*mid, label, color=color, fontsize=12, fontweight='bold')

arrow3(ax1, np.zeros(3), u3, 'royalblue', r'$\mathbf{u}$')
arrow3(ax1, u3, u3+v3, 'firebrick', r'$\mathbf{v}$')
arrow3(ax1, np.zeros(3), w3, 'darkgreen', r'$\mathbf{u}+\mathbf{v}$')
ax1.set_xlabel(r'$x_1$'); ax1.set_ylabel(r'$x_2$'); ax1.set_zlabel(r'$x_3$')
ax1.set_title('3D Vector Addition', fontweight='bold')
ax1.set_xlim(0,3); ax1.set_ylim(0,3); ax1.set_zlim(0,2)

# --- Right: scalar multiplication ---
ax2 = fig.add_subplot(122)
v2 = np.array([1., 1.])
alphas = [2.0, 1.0, 0.5, -1.0, -1.5]
colors = ['darkgreen', 'royalblue', 'orange', 'firebrick', 'purple']
ax2.axhline(0, color='gray', lw=0.5); ax2.axvline(0, color='gray', lw=0.5)
ax2.set_aspect('equal'); ax2.grid(True, alpha=0.3)
ax2.set_xlim(-2.5, 2.5); ax2.set_ylim(-2.5, 2.5)
for alpha, color in zip(alphas, colors):
    av = alpha * v2
    ax2.annotate('', xy=av, xytext=(0,0),
                 arrowprops=dict(arrowstyle='->', color=color, lw=2.2))
    ax2.text(av[0]+0.05, av[1]+0.05, rf'$\alpha={alpha}$', color=color, fontsize=10)
ax2.set_title(r'Scalar multiplication: $\alpha\mathbf{v}$ for various $\alpha$', fontweight='bold')
ax2.set_xlabel(r'$x_1$'); ax2.set_ylabel(r'$x_2$')

plt.tight_layout()
plt.show()
````
