## 3.1 The Dot Product (Scalar Product)

```{admonition} Definition
:class: definition
For any two vectors $\mathbf{u}$ and $\mathbf{v}$:
$$\mathbf{u}\cdot\mathbf{v} = \|\mathbf{u}\|\,\|\mathbf{v}\|\cos\theta(\mathbf{u},\mathbf{v})$$
where $\theta$ is the angle between them ($0 \leq \theta \leq \pi$).
```

The dot product is a **scalar**. It measures the relative orientation of two vectors.

### Key identities

| Situation | Result | Interpretation |
|-----------|--------|----------------|
| $\mathbf{u}\cdot\mathbf{u}$ | $\|\mathbf{u}\|^2$ | Self-product gives squared magnitude |
| $\mathbf{u}\perp\mathbf{v}$ | $\mathbf{u}\cdot\mathbf{v}=0$ | Vectors are orthogonal |
| Parallel, same direction | $\mathbf{u}\cdot\mathbf{v}=\|\mathbf{u}\|\|\mathbf{v}\|$ | $\theta=0$ |
| Parallel, opposite direction | $\mathbf{u}\cdot\mathbf{v}=-\|\mathbf{u}\|\|\mathbf{v}\|$ | $\theta=\pi$ |

### Component form

Since $\hat{\mathbf{e}}_i \cdot \hat{\mathbf{e}}_j = \delta_{ij}$:

$$
\mathbf{u}\cdot\mathbf{v} = u_1v_1 + u_2v_2 + u_3v_3 = \sum_{i=1}^{3} u_i v_i
$$

### Projection

The **projection of $\mathbf{u}$ onto $\mathbf{v}$** (scalar and vector):

$$
\text{proj}_{\mathbf{v}}\mathbf{u} = \frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{v}\|^2}\,\mathbf{v}
$$

````python
"""Interactive: dot product and projection visualisation."""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig, ax = plt.subplots(figsize=(8, 7))
plt.subplots_adjust(bottom=0.35)

def draw_dot(u, v):
    ax.cla()
    norm_u = np.linalg.norm(u); norm_v = np.linalg.norm(v)
    dot = np.dot(u, v)
    cos_theta = dot / (norm_u * norm_v) if (norm_u > 1e-9 and norm_v > 1e-9) else 0.
    cos_theta = np.clip(cos_theta, -1, 1)
    theta_deg = np.degrees(np.arccos(cos_theta))
    proj_vec = (dot / norm_v**2) * v if norm_v > 1e-9 else np.zeros(2)

    lim = max(norm_u, norm_v) + 0.5
    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
    ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)

    # Vectors
    ax.annotate('', xy=u, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='royalblue', lw=2.5))
    ax.text(u[0]+0.1, u[1]+0.1, r'$\mathbf{u}$', color='royalblue', fontsize=14, fontweight='bold')
    ax.annotate('', xy=v, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='firebrick', lw=2.5))
    ax.text(v[0]+0.1, v[1]+0.1, r'$\mathbf{v}$', color='firebrick', fontsize=14, fontweight='bold')

    # Projection of u onto v
    ax.annotate('', xy=proj_vec, xytext=(0,0), arrowprops=dict(arrowstyle='->', color='darkorange', lw=2.0, linestyle='--'))
    ax.plot([u[0], proj_vec[0]], [u[1], proj_vec[1]], 'k:', lw=1.5, label='perpendicular')
    ax.text(proj_vec[0]/2-0.2, proj_vec[1]/2-0.3, r'$\mathrm{proj}_{\mathbf{v}}\mathbf{u}$', color='darkorange', fontsize=10)

    # Angle arc
    angle_v = np.arctan2(v[1], v[0])
    angle_u = np.arctan2(u[1], u[0])
    arc_r = min(norm_u, norm_v) * 0.3
    theta_arc = np.linspace(angle_v, angle_u, 60)
    ax.plot(arc_r*np.cos(theta_arc), arc_r*np.sin(theta_arc), 'k', lw=1.2)
    mid = len(theta_arc)//2
    ax.text(arc_r*1.4*np.cos(theta_arc[mid]), arc_r*1.4*np.sin(theta_arc[mid]), r'$\theta$', fontsize=12)

    # Info box
    info = (f"$\\mathbf{{u}}\\cdot\\mathbf{{v}} = {dot:.3f}$\n"
            f"$\\|\\mathbf{{u}}\\| = {norm_u:.3f}$, $\\|\\mathbf{{v}}\\| = {norm_v:.3f}$\n"
            f"$\\theta = {theta_deg:.1f}^\\circ$\n"
            f"$\\cos\\theta = {cos_theta:.3f}$")
    ax.text(-lim+0.1, lim-0.2, info, fontsize=10, va='top',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))
    fig.canvas.draw_idle()

u0, v0 = np.array([2., 1.]), np.array([1., 2.])
draw_dot(u0, v0)

sliders = {}
positions = [0.08, 0.14, 0.52, 0.58]
labels = [r'$u_1$', r'$u_2$', r'$v_1$', r'$v_2$']
inits = [2., 1., 1., 2.]
axes_s = [plt.axes([0.12, pos, 0.35, 0.025]) for pos in positions[:2]] + \
         [plt.axes([0.55, pos-0.44, 0.35, 0.025]) for pos in positions[2:]]
# Rebuild cleanly:
fig.subplots_adjust(bottom=0.38)
ax_s = [plt.axes([0.12, 0.22, 0.35, 0.025]),
        plt.axes([0.12, 0.15, 0.35, 0.025]),
        plt.axes([0.55, 0.22, 0.35, 0.025]),
        plt.axes([0.55, 0.15, 0.35, 0.025])]
labels2 = [r'$u_1$', r'$u_2$', r'$v_1$', r'$v_2$']
inits2 = [2., 1., 1., 2.]
sliders2 = [Slider(a, lb, -3., 3., valinit=iv, valstep=0.1)
            for a, lb, iv in zip(ax_s, labels2, inits2)]

def update(val):
    draw_dot(np.array([sliders2[0].val, sliders2[1].val]),
             np.array([sliders2[2].val, sliders2[3].val]))

for s in sliders2: s.on_changed(update)
ax.set_title('Interactive Dot Product & Projection', fontsize=12, fontweight='bold')
plt.show()
````

## 3.2 The Cross Product (Vector Product)

```{admonition} Definition
:class: definition
For any two vectors $\mathbf{u}$ and $\mathbf{v}$:
$$\mathbf{u}\times\mathbf{v} = \|\mathbf{u}\|\,\|\mathbf{v}\|\sin\theta\;\hat{\mathbf{n}}$$
where $\hat{\mathbf{n}}$ is the unit normal to the plane of $\mathbf{u}$ and $\mathbf{v}$, determined by the **right-hand rule**.
```

The cross product is a **vector**. Its magnitude equals the area of the parallelogram spanned by $\mathbf{u}$ and $\mathbf{v}$.

### Key properties

- **Anti-commutativity**: $\mathbf{u}\times\mathbf{v} = -(\mathbf{v}\times\mathbf{u})$
- **Parallel vectors**: $\mathbf{u}\times\mathbf{u} = \mathbf{0}$
- **Orthogonal to both**: $(\mathbf{u}\times\mathbf{v})\cdot\mathbf{u} = 0$ and $(\mathbf{u}\times\mathbf{v})\cdot\mathbf{v} = 0$

### Basis vector cross products

$$
\hat{\mathbf{e}}_1\times\hat{\mathbf{e}}_2 = \hat{\mathbf{e}}_3, \qquad
\hat{\mathbf{e}}_2\times\hat{\mathbf{e}}_3 = \hat{\mathbf{e}}_1, \qquad
\hat{\mathbf{e}}_3\times\hat{\mathbf{e}}_1 = \hat{\mathbf{e}}_2
$$

### Component form (determinant formula)

$$
\mathbf{u}\times\mathbf{v} = 
\begin{vmatrix}
\hat{\mathbf{e}}_1 & \hat{\mathbf{e}}_2 & \hat{\mathbf{e}}_3 \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3
\end{vmatrix}
= (u_2v_3-u_3v_2)\hat{\mathbf{e}}_1 - (u_1v_3-u_3v_1)\hat{\mathbf{e}}_2 + (u_1v_2-u_2v_1)\hat{\mathbf{e}}_3
$$

## 3.3 Triple Products

### Scalar triple product

$$
(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w} = 
\begin{vmatrix}
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3 \\
w_1 & w_2 & w_3
\end{vmatrix}
$$

**Geometric meaning**: volume of the parallelepiped spanned by $\mathbf{u}$, $\mathbf{v}$, $\mathbf{w}$.

**Cyclic property**: $(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w} = (\mathbf{v}\times\mathbf{w})\cdot\mathbf{u} = (\mathbf{w}\times\mathbf{u})\cdot\mathbf{v}$

### Vector triple product (BAC–CAB rule)

$$
\mathbf{a}\times(\mathbf{b}\times\mathbf{c}) = \mathbf{b}(\mathbf{a}\cdot\mathbf{c}) - \mathbf{c}(\mathbf{a}\cdot\mathbf{b})
$$

Note: $\mathbf{a}\times(\mathbf{b}\times\mathbf{c}) \neq (\mathbf{a}\times\mathbf{b})\times\mathbf{c}$ in general.


````python
"""Interactive 3D visualisation of the cross product."""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from matplotlib.widgets import Slider

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(bottom=0.3)

def draw_cross(u, v):
    ax.cla()
    w = np.cross(u, v)
    area = np.linalg.norm(w)
    dot_wu = np.dot(w, u); dot_wv = np.dot(w, v)

    # Parallelogram
    verts = [[u*0, u, u+v, v]]
    poly = Poly3DCollection(verts, alpha=0.25, facecolor='gold', edgecolor='goldenrod', linewidth=1.5)
    ax.add_collection3d(poly)

    # u, v vectors
    for vec, color, label in [(u,'royalblue',r'$\mathbf{u}$'), (v,'firebrick',r'$\mathbf{v}$')]:
        ax.quiver(0,0,0, *vec, color=color, linewidth=2.5, arrow_length_ratio=0.1)
        ax.text(*(vec*0.6), label, color=color, fontsize=13, fontweight='bold')

    # Cross product
    scale = min(1.5, 3./max(area, 0.1))
    ax.quiver(0,0,0, *(w/max(area,0.01)*scale), color='darkgreen', linewidth=2.5, arrow_length_ratio=0.12)
    ax.text(*(w/max(area,0.01)*scale*0.6 + np.array([0.1,0.1,0.1])),
            r'$\mathbf{u}\times\mathbf{v}$', color='darkgreen', fontsize=12, fontweight='bold')

    lim = max(np.linalg.norm(u), np.linalg.norm(v), scale) + 0.5
    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim); ax.set_zlim(-lim, lim)
    ax.set_xlabel(r'$x_1$'); ax.set_ylabel(r'$x_2$'); ax.set_zlabel(r'$x_3$')
    ax.set_title(
        f'$|\\mathbf{{u}}\\times\\mathbf{{v}}| = {area:.3f}$ (= area of parallelogram)\n'
        f'$(\\mathbf{{u}}\\times\\mathbf{{v}})\\cdot\\mathbf{{u}} = {dot_wu:.6f}$ (should be 0)   '
        f'$(\\mathbf{{u}}\\times\\mathbf{{v}})\\cdot\\mathbf{{v}} = {dot_wv:.6f}$ (should be 0)',
        fontsize=10)
    ax.view_init(elev=25, azim=40)
    fig.canvas.draw_idle()

u0 = np.array([2., 0., 0.]); v0 = np.array([0., 2., 0.])
draw_cross(u0, v0)

sl_axes = [plt.axes([0.12, pos, 0.30, 0.025])
           for pos in [0.20, 0.15, 0.10]]
sl_axes2 = [plt.axes([0.58, pos, 0.30, 0.025])
            for pos in [0.20, 0.15, 0.10]]
su = [Slider(a, f'$u_{i+1}$', -2.5, 2.5, valinit=u0[i], valstep=0.25)
      for i, a in enumerate(sl_axes)]
sv = [Slider(a, f'$v_{i+1}$', -2.5, 2.5, valinit=v0[i], valstep=0.25)
      for i, a in enumerate(sl_axes2)]

def update(val):
    draw_cross(np.array([s.val for s in su]), np.array([s.val for s in sv]))

for s in su+sv: s.on_changed(update)
plt.show()
````

````python
"""Symbolic verification of the BAC-CAB rule."""
import sympy as sp

a1,a2,a3,b1,b2,b3,c1,c2,c3 = sp.symbols('a1:4 b1:4 c1:4', real=True)
a = sp.Matrix([a1,a2,a3])
b = sp.Matrix([b1,b2,b3])
c = sp.Matrix([c1,c2,c3])

lhs = a.cross(b.cross(c))
rhs = b * a.dot(c) - c * a.dot(b)

diff = sp.simplify(lhs - rhs)
print("BAC-CAB: a × (b × c) = b(a·c) - c(a·b)")
print("Residual (should be zero vector):", diff.T)

# Numerical example
import numpy as np
a_n = np.array([1., 2., 3.])
b_n = np.array([4., 0., -1.])
c_n = np.array([2., -1., 1.])

lhs_n = np.cross(a_n, np.cross(b_n, c_n))
rhs_n = b_n * np.dot(a_n, c_n) - c_n * np.dot(a_n, b_n)
print("\nNumerical check:")
print("  LHS:", lhs_n)
print("  RHS:", rhs_n)
print("  Differ by:", np.linalg.norm(lhs_n - rhs_n))
````
