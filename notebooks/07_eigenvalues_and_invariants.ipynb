## 7.1 Eigenvalues and Eigenvectors

:::{admonition} Definition
:class: note
A scalar $\lambda$ and a nonzero unit vector $\hat{\mathbf{n}}$ are an **eigenvalue–eigenvector pair** of tensor $\mathbf{T}$ if:

$$\mathbf{T}\hat{\mathbf{n}} = \lambda\hat{\mathbf{n}}$$

The eigenvector gives the direction that is **not rotated** by $\mathbf{T}$, only scaled.
:::

### Finding eigenvalues: the characteristic equation

The equation $\mathbf{T}\hat{\mathbf{n}} = \lambda\hat{\mathbf{n}}$ can be rewritten as:

$$(\mathbf{T} - \lambda\mathbf{I})\hat{\mathbf{n}} = \mathbf{0}$$

For a nontrivial solution, the matrix $(\mathbf{T} - \lambda\mathbf{I})$ must be singular:

$$\det(\mathbf{T} - \lambda\mathbf{I}) = 0$$

Expanding the determinant gives the **characteristic polynomial** (cubic in 3D):

$$-\lambda^3 + I_T\lambda^2 - II_T\lambda + III_T = 0$$

where $I_T, II_T, III_T$ are the **principal invariants**:

$$I_T = \text{tr}(\mathbf{T}) = T_{ii}$$

$$II_T = \frac{1}{2}\left[I_T^2 - \text{tr}(\mathbf{T}^2)\right] = \frac{1}{2}\left[(T_{ii})^2 - T_{ij}T_{ji}\right]$$

$$III_T = \det(\mathbf{T})$$

---

## 7.2 Properties for Symmetric Tensors

For a **symmetric** tensor $\mathbf{T} = \mathbf{T}^T$:

- All eigenvalues $\lambda_1, \lambda_2, \lambda_3$ are **real**
- Eigenvectors corresponding to distinct eigenvalues are **mutually orthogonal**
- The eigenvectors $\hat{\mathbf{n}}_1, \hat{\mathbf{n}}_2, \hat{\mathbf{n}}_3$ form a basis for $\mathbb{R}^3$

In the basis of eigenvectors, $\mathbf{T}$ is diagonal:

$$\mathbf{T} = \lambda_1 (\hat{\mathbf{n}}_1 \otimes \hat{\mathbf{n}}_1) + \lambda_2 (\hat{\mathbf{n}}_2 \otimes \hat{\mathbf{n}}_2) + \lambda_3 (\hat{\mathbf{n}}_3 \otimes \hat{\mathbf{n}}_3) = \sum_{i=1}^{3} \lambda_i (\hat{\mathbf{n}}_i \otimes \hat{\mathbf{n}}_i)$$

This is the **spectral decomposition** (or eigendecomposition) of $\mathbf{T}$.

---

## 7.3 Principal Invariants and the Cayley–Hamilton Theorem

The three principal invariants $I_T, II_T, III_T$ are scalar quantities that do **not** change when the coordinate system is changed. They have geometric/physical meaning:

| Invariant | Expression | Physical meaning |
|-----------|-----------|------------------|
| $I_T$ | $\text{tr}(\mathbf{T}) = \lambda_1 + \lambda_2 + \lambda_3$ | Sum of eigenvalues |
| $II_T$ | $\frac{1}{2}[I_T^2 - \text{tr}(\mathbf{T}^2)] = \lambda_1\lambda_2 + \lambda_2\lambda_3 + \lambda_3\lambda_1$ | Sum of products of pairs |
| $III_T$ | $\det(\mathbf{T}) = \lambda_1\lambda_2\lambda_3$ | Product of eigenvalues |

:::{admonition} Cayley–Hamilton Theorem
:class: important
Every tensor satisfies its own characteristic equation:

$$-\mathbf{T}^3 + I_T\mathbf{T}^2 - II_T\mathbf{T} + III_T\mathbf{I} = \mathbf{0}$$

This means that $\mathbf{T}^3$ can always be expressed as a linear combination of $\mathbf{I}$, $\mathbf{T}$, and $\mathbf{T}^2$ — higher powers are never needed!
:::

Note: the fundamental invariants $\{I_T, II_T, III_T\}$ and principal invariants are related but not identical — both appear in the literature.

## 7.4 Interactive: Eigenvalues and Spectral Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, FloatSlider, Layout

def plot_eigen(T11=3.,T12=1.,T13=0.,
               T22=2.,T23=0.,T33=1.):
    # Force symmetric
    T = np.array([[T11,T12,T13],
                  [T12,T22,T23],
                  [T13,T23,T33]])

    eigvals, eigvecs = np.linalg.eigh(T)  # eigh for symmetric
    # Sort by descending eigenvalue
    idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:, idx]

    fig = plt.figure(figsize=(14, 5))

    # ── 3D: eigenvectors ──────────────────────────────────────────
    ax = fig.add_subplot(131, projection='3d')
    ax.set_xlim(-1.5,1.5); ax.set_ylim(-1.5,1.5); ax.set_zlim(-1.5,1.5)
    colors = ['#C0392B','#27AE60','#2980B9']
    for i, (ev, lam, c) in enumerate(zip(eigvecs.T, eigvals, colors)):
        ax.quiver(0,0,0,ev[0],ev[1],ev[2], color=c, linewidth=2.5, arrow_length_ratio=0.15)
        ax.text(ev[0]*1.2,ev[1]*1.2,ev[2]*1.2,
                fr'$\hat{{n}}_{i+1}$ ($\lambda={lam:.3f}$)', fontsize=8, color=c)
    ax.set_xlabel(r'$x_1$'); ax.set_ylabel(r'$x_2$'); ax.set_zlabel(r'$x_3$')
    ax.set_title('Principal Directions', fontsize=11)

    # ── Matrix heatmap ─────────────────────────────────────────────
    ax2 = fig.add_subplot(132)
    vmax = max(np.abs(T).max(), 0.1)
    im = ax2.imshow(T, cmap='RdBu_r', vmin=-vmax, vmax=vmax)
    for i in range(3):
        for j in range(3):
            ax2.text(j,i,f'{T[i,j]:.2f}', ha='center',va='center', fontsize=12,
                     fontweight='bold',
                     color='white' if abs(T[i,j])>0.4*vmax else 'black')
    ax2.set_xticks([0,1,2]); ax2.set_yticks([0,1,2])
    ax2.set_xticklabels(['j=1','j=2','j=3'])
    ax2.set_yticklabels(['i=1','i=2','i=3'])
    ax2.set_title(r'$\mathbf{T}$ (symmetric)', fontsize=11)
    plt.colorbar(im, ax=ax2, fraction=0.046)

    # ── Invariants summary ─────────────────────────────────────────
    ax3 = fig.add_subplot(133)
    ax3.axis('off')
    IT   = np.trace(T)
    IIT  = 0.5*(IT**2 - np.trace(T@T))
    IIIT = np.linalg.det(T)
    ch_check = -T@T@T + IT*T@T - IIT*T + IIIT*np.eye(3)

    rows = [
        ['Quantity','Value','From eigenvalues'],
        [r'$\lambda_1$', f'{eigvals[0]:.5f}','-'],
        [r'$\lambda_2$', f'{eigvals[1]:.5f}','-'],
        [r'$\lambda_3$', f'{eigvals[2]:.5f}','-'],
        [r'$I_T=\mathrm{tr}(T)$', f'{IT:.5f}',
         f'{sum(eigvals):.5f}'],
        [r'$II_T$', f'{IIT:.5f}',
         f'{eigvals[0]*eigvals[1]+eigvals[1]*eigvals[2]+eigvals[2]*eigvals[0]:.5f}'],
        [r'$III_T=\det(T)$', f'{IIIT:.5f}',
         f'{eigvals[0]*eigvals[1]*eigvals[2]:.5f}'],
        ['C-H residual', f'{np.linalg.norm(ch_check):.2e}','(should be ~0)'],
    ]
    t = ax3.table(cellText=rows[1:], colLabels=rows[0],
                  loc='center', cellLoc='center')
    t.auto_set_font_size(False); t.set_fontsize(9); t.scale(1.2, 1.5)
    ax3.set_title('Invariants & Cayley-Hamilton', fontsize=11)

    plt.tight_layout()
    plt.show()

sw = dict(min=-5.,max=5.,step=0.1,layout=Layout(width='250px'))
interact(plot_eigen,
         T11=FloatSlider(3., description='T11 (T12=T21)',**sw),
         T12=FloatSlider(1., description='T12=T21',**sw),
         T13=FloatSlider(0., description='T13=T31',**sw),
         T22=FloatSlider(2., description='T22',**sw),
         T23=FloatSlider(0., description='T23=T32',**sw),
         T33=FloatSlider(1., description='T33',**sw));
```

## 7.5 Interactive: The Characteristic Polynomial

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Layout

def plot_char_poly(T11=3.,T12=1.,T13=0.,T22=2.,T23=0.,T33=1.):
    T = np.array([[T11,T12,T13],[T12,T22,T23],[T13,T23,T33]])
    IT   = np.trace(T)
    IIT  = 0.5*(IT**2 - np.trace(T@T))
    IIIT = np.linalg.det(T)

    lam = np.linspace(min(np.linalg.eigvalsh(T))-1.5,
                      max(np.linalg.eigvalsh(T))+1.5, 500)
    p = -lam**3 + IT*lam**2 - IIT*lam + IIIT
    true_eigs = np.linalg.eigvalsh(T)

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(lam, p, '#2C3E50', lw=2.5, label=r'$p(\lambda) = -\lambda^3+I_T\lambda^2-II_T\lambda+III_T$')
    ax.axhline(0, color='k', lw=0.8)
    for eig in true_eigs:
        ax.axvline(eig, color='#E74C3C', lw=1.5, linestyle='--', alpha=0.7)
        ax.scatter([eig],[0], color='#E74C3C', s=80, zorder=5)
        ax.text(eig, max(p)*0.08, fr' $\lambda={eig:.3f}$',
                color='#E74C3C', fontsize=9, rotation=90, va='bottom')
    ax.set_xlabel(r'$\lambda$', fontsize=12)
    ax.set_ylabel(r'$p(\lambda)$', fontsize=12)
    ax.set_title('Characteristic Polynomial of T', fontsize=12)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

    txt = (fr'$I_T = {IT:.3f}$,  $II_T = {IIT:.3f}$,  $III_T = {IIIT:.3f}$\n'
           fr'Roots (eigenvalues): {sorted(true_eigs, reverse=True)}')
    ax.text(0.02, 0.97, txt, transform=ax.transAxes, fontsize=9,
            va='top', bbox=dict(boxstyle='round', facecolor='#FEF9E7', alpha=0.9))
    plt.tight_layout()
    plt.show()

sw = dict(min=-5.,max=5.,step=0.1,layout=Layout(width='250px'))
interact(plot_char_poly,
         T11=FloatSlider(3.,description='T11',**sw),
         T12=FloatSlider(1.,description='T12=T21',**sw),
         T13=FloatSlider(0.,description='T13=T31',**sw),
         T22=FloatSlider(2.,description='T22',**sw),
         T23=FloatSlider(0.,description='T23=T32',**sw),
         T33=FloatSlider(1.,description='T33',**sw));
```

## 7.6 Module Summary

We have covered the complete algebra of vectors and tensors in 3D Euclidean space:

| Topic | Key idea |
|-------|----------|
| Vectors | Magnitude + direction; coordinate-independent |
| Addition, scaling | Head-to-tail; component-wise |
| Dot product | Projects one vector onto another; gives angle |
| Cross product | Area of parallelogram; right-hand rule |
| Index notation | Economy + power: Einstein convention, $\delta_{ij}$, $\varepsilon_{ijk}$ |
| Matrices / rotations | $[Q]$ orthogonal; $\det=\pm1$; $[Q]^{-1}=[Q]^T$ |
| Tensors | Linear maps; two indices; transform with two $Q$'s |
| Eigenvalues | $\mathbf{T}\hat{n}=\lambda\hat{n}$; real & orthogonal for symmetric tensors |
| Spectral decomp. | $\mathbf{T}=\sum_i\lambda_i(\hat{n}_i\otimes\hat{n}_i)$ |
| Invariants | $I_T, II_T, III_T$: coordinate-independent scalars |
| Cayley–Hamilton | Every tensor satisfies its own characteristic equation |

The Cauchy stress tensor at a point in a structure is:
$$[\boldsymbol{\sigma}] = \begin{bmatrix}100 & 40 & 0\\ 40 & -20 & 0 \\ 0 & 0 & 60\end{bmatrix} \text{ MPa}$$
   - Find all principal stresses (eigenvalues) and principal directions (eigenvectors).
   - Compute the three invariants $I_{\sigma}, II_{\sigma}, III_{\sigma}$.
   - Verify that Cayley–Hamilton holds.

```python
import numpy as np

# 7.1 verification
T = np.array([[5.,2.,0.],[2.,5.,0.],[0.,0.,3.]])
eigvals, eigvecs = np.linalg.eigh(T)
print("7.1  T ="); print(T)
print(f"     eigenvalues: {sorted(eigvals, reverse=True)}")
print("     eigenvectors (columns):")
print(eigvecs)

# Spectral reconstruction
T_spectral = sum(lam * np.outer(v,v) for lam,v in zip(eigvals, eigvecs.T))
print(f"     Spectral reconstruction matches T: {np.allclose(T, T_spectral)}")

# Cayley-Hamilton
IT   = np.trace(T)
IIT  = 0.5*(IT**2 - np.trace(T@T))
IIIT = np.linalg.det(T)
ch   = -T@T@T + IT*T@T - IIT*T + IIIT*np.eye(3)
print(f"     Cayley-Hamilton residual ||..|| = {np.linalg.norm(ch):.2e}  (should be ~0)")

print()
print("7.3  Diagonal tensor invariants:")
T3 = np.diag([1.,2.,3.])
print(f"     I_T = tr = {np.trace(T3):.0f}  (= 1+2+3)")
print(f"     III_T = det = {np.linalg.det(T3):.0f}  (= 1x2x3)")

print()
print("Tutorial 7.4 - Principal Stress Analysis:")
sigma = np.array([[100.,40.,0.],[40.,-20.,0.],[0.,0.,60.]])
lams, nvecs = np.linalg.eigh(sigma)
lams_sorted = sorted(lams, reverse=True)
print(f"  Principal stresses: {[round(l,4) for l in lams_sorted]} MPa")
print(f"  I_sigma  = {np.trace(sigma):.4f} MPa")
print(f"  III_sigma = det = {np.linalg.det(sigma):.4f} MPa^3")
tau_max = (max(lams)-min(lams))/2
print(f"  tau_max = (lambda_max - lambda_min)/2 = {tau_max:.4f} MPa")
```
