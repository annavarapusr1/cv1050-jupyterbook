## 5.1 Matrices as Linear Maps

An $m\\times n$ matrix $\\mathbf{A}$ stores $mn$ real numbers $A_{ij}$. In our context, matrices are the **component representations** of linear maps (tensors) in a chosen basis.

### Key operations in index notation

| Operation | Index form | Matrix form |
|-----------|-----------|-------------|
| Matrix–vector product | $b_i = A_{ij}x_j$ | $\\mathbf{b}=\\mathbf{A}\\mathbf{x}$ |
| Matrix–matrix product | $[\\mathbf{AB}]_{ij} = A_{ik}B_{kj}$ | $\\mathbf{AB}$ |
| Transpose | $[\\mathbf{A}^T]_{ij} = A_{ji}$ | — |
| Trace | $\\text{tr}(\\mathbf{A}) = A_{ii}$ | Sum of diagonal |
| Frobenius product | $\\mathbf{A}:\\mathbf{B} = A_{ij}B_{ij}$ | — |

### Symmetric/skew decomposition

Any matrix decomposes uniquely as:

$$\\mathbf{A} = \\underbrace{\\frac{\\mathbf{A}+\\mathbf{A}^T}{2}}_{\\mathbf{S}\\text{ (symmetric)}} + \\underbrace{\\frac{\\mathbf{A}-\\mathbf{A}^T}{2}}_{\\mathbf{W}\\text{ (skew-symmetric)}}$$

$\\mathbf{S}$ has 6 independent components; $\\mathbf{W}$ has 3.

## 5.2 Coordinate Transformation

Consider two right-handed orthonormal frames sharing the same origin:

- **Frame** $\\mathcal{F}$: basis $\\{\\hat{\\mathbf{e}}_1,\\hat{\\mathbf{e}}_2,\\hat{\\mathbf{e}}_3\\}$, vector components $\\{v_i\\}$
- **Frame** $\\mathcal{F}'$: basis $\\{\\hat{\\mathbf{e}}'_1,\\hat{\\mathbf{e}}'_2,\\hat{\\mathbf{e}}'_3\\}$, vector components $\\{v'_i\\}$

The **transformation matrix** (rotation matrix) has entries:

$$Q_{ij} = \\hat{\\mathbf{e}}'_i \\cdot \\hat{\\mathbf{e}}_j = \\cos\\angle(\\hat{\\mathbf{e}}'_i, \\hat{\\mathbf{e}}_j)$$

These are the **direction cosines**. The component transformation laws are:

$$\\boxed{v'_i = Q_{ij}v_j} \\qquad \\boxed{v_i = Q_{ji}v'_j}$$

### Orthogonality of Q

Since both frames are orthonormal:

$$Q_{ij}Q_{ik} = \\delta_{jk} \\quad\\Rightarrow\\quad \\mathbf{Q}\\mathbf{Q}^T = \\mathbf{I}, \\quad \\mathbf{Q}^{-1}=\\mathbf{Q}^T, \\quad \\det(\\mathbf{Q})=\\pm 1$$

Proper rotations ($\\det=+1$) form the group $SO(3)$.

### 2D rotation by angle $\\theta$ (anticlockwise)

$$\\mathbf{Q}(\\theta) = \\begin{pmatrix}\\cos\\theta & \\sin\\theta \\\\ -\\sin\\theta & \\cos\\theta\\end{pmatrix}$$

### Tensor (2nd-order) transformation law

A second-order tensor $\\mathbf{T}$ with components $T_{ij}$ in $\\mathcal{F}$ has components in $\\mathcal{F}'$:

$$T'_{ij} = Q_{ik}Q_{jl}T_{kl} \\quad\\Leftrightarrow\\quad [\\mathbf{T}'] = [\\mathbf{Q}][\\mathbf{T}][\\mathbf{Q}]^T$$

Scalar invariants (trace, determinant) are unchanged by this transformation.

```python
"""Interactive: rotate a frame and observe how vector components transform."""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig, ax = plt.subplots(figsize=(8, 7))
plt.subplots_adjust(bottom=0.22)

v_fixed = np.array([2., 1.])  # physical vector fixed in space

def Q2d(theta):
    c, s = np.cos(theta), np.sin(theta)
    return np.array([[c, s], [-s, c]])

def draw(theta_deg):
    theta = np.radians(theta_deg)
    Q = Q2d(theta)
    v_prime = Q @ v_fixed

    ax.cla()
    ax.set_xlim(-3, 3); ax.set_ylim(-3, 3)
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)

    # Original basis
    for vec, lbl in [(np.array([1,0]), r'$\hat{\mathbf{e}}_1$'),
                     (np.array([0,1]), r'$\hat{\mathbf{e}}_2$')]:
        ax.annotate('', xy=2*vec, xytext=(0,0),
                    arrowprops=dict(arrowstyle='->', color='royalblue', lw=1.8))
        ax.text(*(2.1*vec + 0.05), lbl, color='royalblue', fontsize=12)

    # Rotated basis
    e1p = np.array([np.cos(theta), np.sin(theta)])
    e2p = np.array([-np.sin(theta), np.cos(theta)])
    for vec, lbl in [(e1p, r"$\hat{\mathbf{e}}'_1$"), (e2p, r"$\hat{\mathbf{e}}'_2$")]:
        ax.annotate('', xy=2*vec, xytext=(0,0),
                    arrowprops=dict(arrowstyle='->', color='firebrick', lw=1.8))
        ax.text(*(2.15*vec + 0.05), lbl, color='firebrick', fontsize=12)

    # The fixed vector v
    ax.annotate('', xy=v_fixed, xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color='black', lw=2.8))
    ax.text(v_fixed[0]+0.1, v_fixed[1]+0.1, r'$\mathbf{v}$ (unchanged)', fontsize=11, fontweight='bold')

    info = (f"In $\\mathcal{{F}}$:  $v_1={v_fixed[0]:.2f}$, $v_2={v_fixed[1]:.2f}$\n"
            f"In $\\mathcal{{F}}'$: $v'_1={v_prime[0]:.3f}$, $v'_2={v_prime[1]:.3f}$\n"
            f"$\\|\\mathbf{{v}}\\|={np.linalg.norm(v_fixed):.4f}$ = "
            f"$\\|\\mathbf{{v}}'\\|={np.linalg.norm(v_prime):.4f}$ (invariant ✓)")
    ax.text(-2.9, -2.8, info, fontsize=10, va='bottom',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))
    ax.set_title(f'Frame rotated by θ = {theta_deg:.1f}° | Components change, vector does not',
                 fontsize=11, fontweight='bold')
    fig.canvas.draw_idle()

draw(30.)
ax_s = plt.axes([0.15, 0.08, 0.70, 0.03])
sl = Slider(ax_s, r'$\theta$ (degrees)', 0., 180., valinit=30., valstep=1.)
sl.on_changed(lambda v: draw(v))
plt.show()
```

```python
"""Verify rotation matrix properties and tensor transformation."""
import numpy as np

def rotation_3d(axis, angle_deg):
    """Rotation matrix about axis by angle (Rodrigues formula)."""
    theta = np.radians(angle_deg)
    n = np.array(axis, dtype=float); n /= np.linalg.norm(n)
    c, s = np.cos(theta), np.sin(theta)
    K = np.array([[0,-n[2],n[1]],[n[2],0,-n[0]],[-n[1],n[0],0]])
    return np.eye(3) + s*K + (1-c)*(K@K)

Q = rotation_3d([0,0,1], 45.)
print("Q (45° about z):"); print(np.round(Q,4))
print("Q Q^T ="); print(np.round(Q@Q.T, 6))
print(f"det(Q) = {np.linalg.det(Q):.6f}")

v = np.array([2., 1., 0.])
vp = Q @ v
print(f"\n||v|| = {np.linalg.norm(v):.6f}  ||v'|| = {np.linalg.norm(vp):.6f}  (invariant ✓)")

# Tensor transformation: A' = Q A Q^T
A = np.array([[3.,1.,0.],[1.,2.,0.],[0.,0.,5.]])
Ap = Q @ A @ Q.T
print(f"\ntr(A) = {np.trace(A):.4f}   tr(A') = {np.trace(Ap):.4f}  (invariant ✓)")
print(f"det(A) = {np.linalg.det(A):.4f}  det(A') = {np.linalg.det(Ap):.4f}  (invariant ✓)")
```
