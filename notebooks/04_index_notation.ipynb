## 4.1 Einstein Summation Convention

Repeated summations in vector algebra can be written compactly using the **Einstein summation convention**: a repeated index in a single term implies summation over its range.

$$\mathbf{v} = v_i \hat{\mathbf{e}}_i \equiv \sum_{i=1}^{3} v_i \hat{\mathbf{e}}_i$$

| Index type | Appears | Meaning |
|-----------|---------|--------|
| **Dummy** | exactly **twice** in a term | Sum over 1 to $n_{sd}$ |
| **Free** | exactly **once** per term | Stands for $n_{sd}$ separate equations |

**Rules:**
- A free index must appear on **both sides** of every equation
- A dummy index must appear **exactly twice** in a term (not three or more times)
- Dummy indices can be renamed freely (dummy relabelling)

### Examples

$$\mathbf{u}\cdot\mathbf{v} = u_i v_i = u_1v_1 + u_2v_2 + u_3v_3$$

$$\|\mathbf{v}\|^2 = v_i v_i$$

$$v_i = \mathbf{v}\cdot\hat{\mathbf{e}}_i \quad\text{(free index } i\text{: three equations)}$$

---

## 4.2 The Kronecker Delta

$$\delta_{ij} = \hat{\mathbf{e}}_i \cdot \hat{\mathbf{e}}_j = \begin{cases}1 & i=j \\ 0 & i\neq j\end{cases}$$

**Substitution property:** $\delta_{ij} u_j = u_i$ (replaces dummy $j$ with free $i$)

**Trace:** $\delta_{ii} = 3$ (in 3D)

**Product:** $\delta_{ij}\delta_{jk} = \delta_{ik}$

---

## 4.3 The Levi-Civita Symbol

$$\varepsilon_{ijk} = \begin{cases}+1 & (i,j,k) \text{ even perm. of } (1,2,3) \\ -1 & \text{odd perm.} \\ 0 & \text{any repeated index}\end{cases}$$

**Cross product:** $[\mathbf{u}\times\mathbf{v}]_k = \varepsilon_{ijk} u_i v_j$

**The epsilon-delta identity:**

$$\varepsilon_{ijk}\varepsilon_{imn} = \delta_{jm}\delta_{kn} - \delta_{jn}\delta_{km}$$

This identity allows any vector triple product identity to be proved purely algebraically.

## 4.4 Interactive: The Index Objects Visualised

```python
import numpy as np
import matplotlib.pyplot as plt

def levi_civita_3d():
    eps = np.zeros((3,3,3))
    eps[0,1,2]=eps[1,2,0]=eps[2,0,1]=1
    eps[0,2,1]=eps[2,1,0]=eps[1,0,2]=-1
    return eps

eps = levi_civita_3d()
delta = np.eye(3)

fig, axes = plt.subplots(1, 4, figsize=(15, 4))

# Kronecker delta
ax = axes[0]
im = ax.imshow(delta, cmap='RdBu_r', vmin=-1, vmax=1)
for i in range(3):
    for j in range(3):
        ax.text(j,i,f'{int(delta[i,j])}', ha='center', va='center',
                fontsize=16, fontweight='bold',
                color='white' if delta[i,j]!=0 else 'black')
ax.set_xticks([0,1,2]); ax.set_yticks([0,1,2])
ax.set_xticklabels(['j=1','j=2','j=3'])
ax.set_yticklabels(['i=1','i=2','i=3'])
ax.set_title(r'$\delta_{ij}$', fontsize=14)
plt.colorbar(im, ax=ax, fraction=0.046)

# Levi-Civita for k=1,2,3
for k in range(3):
    ax2 = axes[k+1]
    data = eps[:,:,k]
    im2 = ax2.imshow(data, cmap='RdBu_r', vmin=-1, vmax=1)
    for i in range(3):
        for j in range(3):
            v = int(data[i,j])
            ax2.text(j,i, f'{v:+d}' if v!=0 else '0',
                     ha='center', va='center', fontsize=14, fontweight='bold',
                     color='white' if v!=0 else 'gray')
    ax2.set_xticks([0,1,2]); ax2.set_yticks([0,1,2])
    ax2.set_xticklabels(['j=1','j=2','j=3'])
    ax2.set_yticklabels(['i=1','i=2','i=3'])
    ax2.set_title(fr'$\varepsilon_{{ij{k+1}}}$ (k={k+1})', fontsize=14)
    plt.colorbar(im2, ax=ax2, fraction=0.046)

plt.suptitle('The Fundamental Index Objects', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

## 4.5 Numerical Verification of Identities

```python
import numpy as np

def levi_civita():
    e=np.zeros((3,3,3))
    e[0,1,2]=e[1,2,0]=e[2,0,1]=1
    e[0,2,1]=e[2,1,0]=e[1,0,2]=-1
    return e

eps=levi_civita(); delta=np.eye(3)

print('=== Index Notation Identities ===')
print(f'1. delta_ii = {np.einsum("ii",delta):.0f}  (expected 3)')
print(f'2. delta_ij delta_ij = {np.einsum("ij,ij->",delta,delta):.0f}  (expected 3)')
print(f'3. eps_ijk eps_ijk = {np.einsum("ijk,ijk->",eps,eps):.0f}  (expected 6)')
print(f'4. delta_ij eps_ijk = {np.einsum("ij,ijk->k",delta,eps)}  (expected [0,0,0])')

# Epsilon-delta identity
lhs=np.einsum('ijk,imn->jkmn',eps,eps)
rhs=np.einsum('jm,kn->jkmn',delta,delta)-np.einsum('jn,km->jkmn',delta,delta)
print(f'5. eps-delta identity holds: {np.allclose(lhs,rhs)}')

# Cross product via epsilon vs numpy
u=np.array([1.,2.,3.]); v=np.array([4.,-1.,2.])
cross_eps=np.einsum('ijk,j,k->i',eps,u,v)
cross_np=np.cross(u,v)
print(f'6. u x v via eps_ijk: {cross_eps}')
print(f'   u x v via np.cross: {cross_np}')
print(f'   Match: {np.allclose(cross_eps,cross_np)}')

# BAC-CAB
a=np.array([1.,0.,2.]); b=np.array([3.,1.,-1.]); c=np.array([0.,2.,1.])
lhs_bac=np.cross(a,np.cross(b,c))
rhs_bac=b*np.dot(a,c)-c*np.dot(a,b)
print(f'7. BAC-CAB: LHS={lhs_bac}, RHS={rhs_bac}, Equal={np.allclose(lhs_bac,rhs_bac)}')
```
